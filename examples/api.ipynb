# Quickstart with PyMKF
To have the right start at full gas let's see how many methods are available in PyMKF.

```python
from pprint import pprint
import PyMKF

methods = dir(PyMKF)
pprint(methods)
```

The list of the methods is really extensive. Let's split them into groups:

```python
from collections import defaultdict

# Group methods by their prefixes
grouped_methods = defaultdict(list)
for method in methods:
    prefix = method.split('_')[0]
    grouped_methods[prefix].append(method)

# Print grouped methods
for prefix, method_list in grouped_methods.items():
    print(f"{prefix}:")
    pprint(method_list)
    print()
```
